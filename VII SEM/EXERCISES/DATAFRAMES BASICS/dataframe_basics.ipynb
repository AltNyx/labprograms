# Pandas DataFrame

```python
import pandas as pd
```

```python
# Manually creating DF
weather_data = {
    'day': ['1/1/2017', '1/2/2017', '1/4/2017'],
    'temp': [32, 35, 36],
    'windspeed': [6, 7, 2],
    'event': ['Rain', 'Sunny', 'Snow']
}

df = pd.DataFrame(weather_data)
df
```
